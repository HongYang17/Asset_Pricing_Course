```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings 
warnings.filterwarnings("ignore")
```

```python
df1=\
    pd.read_excel("Market_Portfolio.xlsx")
df2=\
    pd.read_excel("Industry_Portfolios.xlsx")
```

```python
df=\
    pd.merge(df1,df2)
```

>`Create a table showing the intercept and slope coefficients for the ten industry portfolios.`

```python
from sklearn.linear_model import LinearRegression
```

```python
Rf=0.13
```

```python
df["Excess_Market_Return"]=\
    df["Market"]-Rf
for col in df.columns[2:]:
    df[f"Excess_{col}_Return"]=df[col]-Rf
results=\
    pd.DataFrame(columns=["Industry","Alpha","Beta"])

for industry in df.columns[2:12]:
    merged_data=\
        df[["Date",f"Excess_{industry}_Return","Excess_Market_Return"]]
    y=merged_data[f"Excess_{industry}_Return"]
    x=merged_data[["Excess_Market_Return"]]
    
    model=LinearRegression()
    model.fit(x,y)
    
    alpha=model.intercept_
    beta=model.coef_[0]
    new_row=\
        pd.DataFrame([{"Industry":industry,"Alpha":alpha,"Beta":beta}])
    results = pd.concat([results, new_row], ignore_index=True)                   
```

```python
results
```

>`Briefly explain (in words, without mathematical equations or formulas) the economic significance and pricing implications of the intercept and slope coefficients.`

<p style="line-height:2.0;">
The intercept(Alpha)  indicates whether the asset or portfolio has outperformed or underperformed the market, adjusting for risk. A positive alpha suggests outperformance, while a negative alpha indicates underperformance.
The slope(Beta)  measures how sensitive the asset or portfolio’s returns are to changes in the broader market, showing how much the asset’s returns are expected to fluctuate with market movements.
</p>

>`Also plot the positions of the ten industry portfolios and the market portfolio. `


$$ E(R_i)=R_f+(E(R_m)-R_f)  \times  \beta_i
$$

$$ \text{Mean Return}=\alpha+\beta \times (Beta)
$$

```python
market_beta = \
    pd.DataFrame({'Industry': ['Market'], 'Beta': [1]})
results = \
    pd.concat([results, market_beta], ignore_index=True)
results
```

```python
x=results[["Beta"]]
y=df5.values

sml_model=LinearRegression()
sml_model.fit(x,y)

sml_intercept=sml_model.intercept_
sml_slope=sml_model.coef_[0]

beta_range=\
    np.linspace(0,2,100)
sml_returns = sml_intercept + sml_slope * beta_range

plt.figure(figsize=(8,6))
plt.plot(beta_range,
         sml_returns,
         label="SML")

plt.scatter(results["Beta"],
            df5["Mean_return"])
plt.xlabel(r'$\beta_i$(Systematic Risk)')
plt.ylabel(r"$R_i$(Expected Return)")
plt.legend()

plt.show()
```

>`Briefly explain the economic significance and pricing implications of the SML.`

<p style="line-height:2.0;">
    The SML illustrates the trade-off between risk and return in financial markets, evaluating whether an asset or portfolio is fairly priced based on its risk—assets below the SML are considered overpriced, while those above it are considered underpriced. It shows how much return investors require for taking on different levels of market (systematic) risk, with beta representing an asset's sensitivity to overall market movements. Investors can use the SML to decide whether to buy, sell, or hold an asset based on its position relative to the SML. Assets above the line may be potential buys, while those below it might be considered for selling.
