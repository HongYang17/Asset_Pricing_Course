#### Regress the monthly excess returns for each industry portfolio on the three Fama–French risk factors:

>Create a table showing the factor loadings on SMB and HML for the ten industry portfolios.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import statsmodels.api as sm
import warnings 
warnings.filterwarnings("ignore")
```

```python
df1=\
    pd.read_excel("Risk_Factors.xlsx")
df2=\
    pd.read_excel("Market_Portfolio.xlsx")
df3=\
    pd.read_excel("Industry_Portfolios.xlsx")
```

```python
for i in df3.columns[1:]:
    df1[f"Excess_{i}_return"]=df3[i]-df1["Rf"]
```

```python
Rf=df1["Rf"]
MKT=df1["Rm-Rf"]
SMB=df1["SMB"]
HML=df1["HML"]
```

$$
R_i = \alpha + \beta_{\text{MKT-RF}} \cdot \text{(MKT-RF)} + \beta_{\text{SMB}} \cdot \text{SMB} + \beta_{\text{HML}} \cdot \text{HML} + \epsilon
$$

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X=\
(
    pd.DataFrame({
        "MKT-RF":MKT,"SMB":SMB,"HML":HML
    })
)
Ri=df1.iloc[:,5:]
model.fit(X, Ri)
coefficients=model.coef_
intercepts=model.intercept_
results_df=pd.DataFrame(coefficients,columns=["MKT-RF loading","SMB loading","HML loading"])
results_df["α"]=intercepts
results_df.index=Ri.columns
results_df
```

#### Using monthly excess returns for the ten industry portfolios, calculate the following performance metrics:

>Sharpe Ratio

$$
S_i = \frac{E\left( \tilde{R}_i - R_f \right)}{\sqrt{\text{Var}\left( \tilde{R}_i - R_f \right)}}
$$
$
$


```python
average_return=Ri.mean()
std=Ri.std()
Sharpe_ratio=average_return/std
```

>Below-target semi-variance

$$
SV\left( \tilde{R}_i; \tilde{R}_t \right) = E \left[ \min \left\{ \tilde{R}_i - \tilde{R}_t, 0 \right\}^2 \right]
$$


```python
below_target=(np.minimum(Ri, 0) ** 2).mean()
```

>Sortino Ratio

$$
St_i = \frac{E\left( \tilde{R}_i - \tilde{R}_t \right)}{\sqrt{SV\left( \tilde{R}_i; \tilde{R}_t \right)}}
$$
$
$


```python
below_target_std=np.sqrt(below_target)
sortino_ratio=average_return/below_target_std
```

```python
model1 = LinearRegression()
X1=\
(
    pd.DataFrame({
        "MKT-RF":MKT
    })
)
Ri=df1.iloc[:,5:]
model1.fit(X1, Ri)
coefficients1=model1.coef_
intercepts1=model1.intercept_
results1=pd.DataFrame(coefficients1,columns=["β"])
results1["α1"]=intercepts
results1.index=Ri.columns
results1
```

```python
β=results1["β"]
```

>Treynor Ratio

$$
T_i = \frac{E\left( \tilde{R}_i - R_f \right)}{\beta_i}
$$

```python
Treynor_ratio=average_return/β
```

```python
average_MKT=MKT.mean()
```

>Jensen's α

$$
\alpha_i = E\left( \tilde{R}_i - R_f \right) - \beta_i E\left( \tilde{R}_m - R_f \right)
$$
$
$


```python
Jensen_α=average_return-β*average_MKT
```

```python
data = {
    "Sharpe ratio": Sharpe_ratio,  
    "Sortino ratio": sortino_ratio,    
    "Treynor ratio": Treynor_ratio,
    "Jensen's α":Jensen_α,
    "Fama–French three-factor α":intercepts
}
result_series = pd.DataFrame(data)
result_series
```

>Briefly explain (in words, without mathematical equations or formulas) the economic significance and pricing implications of each of the three performance ratios (but not α's).

1.Sharpe Ratio: Assesses the return generated by an investment for each unit of total risk. A higher Sharpe Ratio indicates more efficient risk-adjusted performance, making the investment more appealing.

2.Sortino Ratio: Concentrates solely on downside risk, emphasizing how well an investment minimizes losses while delivering positive returns. It is particularly relevant for risk-averse investors. A higher Sortino Ratio suggests greater returns with lower downside risk.

3.Treynor Ratio: Measures returns in relation to market risk (systematic risk). A higher Treynor Ratio reflects better compensation for market exposure, indicating that the investment has achieved higher returns with lower exposure to market risk.
