### Plot the minimum-variance frontier (without the riskless asset) generated by the ten industry portfolios

> This graph must have expected (monthly) return on the vertical axis vs standard deviation of (monthly) return on the horizontal axis.\
> This graph must cover the range from 0% to 2% on the vertical axis, in increments of 0.1% (or less).

$$ σ^2=w^TVw $$

```python
cov_matrix=df_1.cov()
cov_matrix
```

```python
R=mean_return.values
V=cov_matrix.values
e=np.ones(len(df_1.columns)).T
V_inv=np.linalg.inv(V)
```

$$ \alpha=R'V^{-1} e $$

```python
alpha=np.dot(np.dot(R.T,V_inv),e)
```

```python
alpha
```

$$ \zeta=R' V^{-1} R $$

```python
zeta=np.dot(np.dot(R.T,V_inv),R)
```

```python
zeta
```

$$ \delta=e' V^{-1} e $$

```python
delta=np.dot(np.dot(e.T,V_inv),e)
```

```python
delta
```

$$ σ_p^2=\frac{1}{\delta} +\frac{\delta}{\zeta \delta - \alpha^2}(R_p - R_{mv})^2 $$

```python
R_p=np.arange(0,2,0.1)
R_mv=alpha/delta
R_mv
```

```python
R_p
```

```python
sigma_p=\
    (1/delta)+(delta/(zeta*delta-alpha**2))*(R_p-R_mv)**2
sigma_p_squared=\
    np.sqrt(sigma_p)
```

```python
sigma_p_squared
```

```python
plt.plot(sigma_p_squared,R_p,label="Minimun Variance Frontier")
plt.xticks(np.arange(2.5, 5, 0.25))
plt.xlabel(r"$\sigma_p$") 
plt.ylabel("Expected Return (Rp)")
plt.title(r"Rp vs. $\sigma_p$")
plt.legend
plt.grid(True)

plt.show()
```

> Briefly explain (in words, without mathematical equations or formulas) the economic significance and relevance of the minimum-variance frontier to an investor

`The minimum_variance frontier illustrates the set of portfolios that offer the lowest level of risk (volatility) for a given level of expected return. According to investors, they should maximize their returns and minimize their risk at the same time.The inflection point on the left side of the chart represents the minimum variance portfolio, that is, the portfolio with the least risk among all possible portfolios. At this point, investors can achieve a certain gain with minimal volatility (minimal risk).The frontier helps to find the optimal protfolios. So they only need to take the upper half of the curve into account, which is regarded as efficient froniter. Besides, by combining different assets, diversity can reduce the unsystematic risks.`

### Now suppose that the (net) risk-free rate is 0.13% per month:

> Plot the efficient frontier (with the riskless asset) on the same graph as the minimum-variance frontier generated by the ten industry portfolios.

$$ \sigma_p^2=\frac{(R_p-R_f)^2}{\zeta-2\alpha R_f+ \delta R_f^2} $$

```python
R_f=0.13
```

```python
sigma_p_sqr=(R_p-R_f)**2/(zeta-2*alpha*R_f+ delta*(R_f)**2)
```

```python
sigma_p_sqr
```

```python
sigma_p1=sigma_p_sqr**0.5
sigma_p1
```

$$R_{p} = R_f + \sqrt{\left( \zeta - 2 \alpha R_f + \delta R_f^2 \right)}\sigma_{p}
$$

```python
R_p1=R_f+(zeta-2*alpha*R_f+delta*(R_f)**2)**0.5*sigma_p1
```

```python
R_p1
```

```python
plt.plot(sigma_p_squared,R_p,label="Minimum_variance frontier")
plt.plot(sigma_p1, R_p1,label="Efficient frontier")
plt.title(r"Rp vs. $\sigma_p$ with riskless asset")
plt.ylabel("Expected Return(R_p)")
plt.xlabel(r"$\sigma_p$")
plt.legend()
plt.grid(True)
plt.show()
```

> Briefly explain the economic significance and relevance of the efficient frontier to an investor.

`The minimum-variance frontier shows how to combine risky assets to minimize risk for a given level of return. However, the CML(orange line) represents the most efficient portfolios you can achieve by mixing a risk-free asset with the market portfolio. Any point on the CML offers the best possible return for the risk taken, outperforming portfolios that only include risky assets. The goal is to find the optimal mix of risky assets and the risk-free asset based on your risk tolerance, allowing you to maximize returns while managing risk efficiently.`

Economic significance:

Minimum-variance frontier represents outermost envelope of attainable portfolios ⇒ no attainable portfolio exists to the left of the minimum-variance frontier.

Efficient frontier consists of portfolios with highest mean return for specified standard deviation of return ⇒ risk-averse investor will invest in (optimal) portfolio on efficient frontier that maximises expected utility (of wealth).`trader option`

Tangency portfolio has the highest possible Sharpe ratio.

### The two frontiers will intersect at single point: the tangency portfolio
>
Calculate the Sharpe ratio for the tangency portfolio, and also the tangency portfolio weights for the ten industry portfolios.

$$ R_{tg}=\frac{\alpha R_f-\zeta}{\delta R_f-\alpha} $$

```python
R_tg=(alpha*R_f-zeta)/(delta*R_f-alpha)
R_tg
```

$$ 
\frac{R_{tg} - R_f}{\sigma_{tg}} = \left( \zeta - 2 \alpha R_f + \delta (R_f)^2 \right)^{\frac{1}{2}}
$$

```python
SP_ratio= (zeta-2*alpha*R_f+delta*(R_f)**2)**0.5
print(colored(f"Sharp Ratio = {SP_ratio}", "black", "on_yellow"))
```

$$ \sigma_{tg} = \frac{-\left( \zeta - 2 \alpha R_f + \delta R_f^2 \right)^{\frac{1}{2}}}{\delta \left( R_f - R_{mv} \right)}  $$

```python
sigma_tg=-(zeta-2*alpha*R_f+delta*R_f**2)**0.5/(delta*(R_f-R_mv))
sigma_tg
```

$$ \alpha=\frac{\zeta V^{-1}e-\alpha V^{-1}R}{\zeta \delta-\alpha^2} $$

```python
a=[zeta*np.dot(V_inv,e)-alpha*np.dot(V_inv,R)]/(zeta*delta-alpha**2)
a
```

$$ b=\frac{\delta V^{-1}R-\alpha V^{-1}e}{\zeta\delta-\alpha^{2}} $$

```python
b=[delta*np.dot(V_inv,R)-alpha*np.dot(V_inv,e)]/(zeta*delta-alpha**2)
b
```

$$ w^{*}=a + bR_p $$
$$ (R_p=R_{tg}) $$

```python
w_s=a+np.dot(b,R_tg)
print(colored(f"w_s = {w_s}", "black", "on_yellow"))
```

```python
pd.DataFrame(w_s.T,df_1.columns,columns=["Weights"])
```

> Briefly explain the economic significance and relevance of the tangency portfolio to an investor.

`At the tangent point, the portfolio has the highest Sharpe Ratio, that is, the highest excess return per unit of risk. This is the optimal risk-return mix.
The closer to the upper right of the CML,CRRA investors are willing to take on more risk, so they allocate more wealth to risky assets, and may even achieve higher expected returns by borrowing to invest in a larger portfolio of markets.
The portfolio weights show the specific allocation of investors between different risk assets, with positive weights indicating a bullish position and negative weights indicating a bearish or hedged short position.`
