Consumption growth has lognormal distribution:

$$ \ln \tilde{g} = 0.02 + 0.02 \tilde{\epsilon}
$$

where ε is standard normal random variable. Simulate probability distribution for consumption growth with (at least) 10^4 random draws from standard normal distribution.

$$
\ln \tilde{g} = \mu + \sigma \tilde{\epsilon}
$$

```python
n_simulations=10**4
mu_g=0.02
sigma_g=0.02
```

```python
epsilon_tilde=np.random.normal(0,1,n_simulations)
ln_g_tilde=mu_g+sigma_g*epsilon_tilde
g_tilde=np.exp(ln_g_tilde)
g_tilde.shape[0]
```

With these parameters, risk-free rate is around 3% per year:

$$
R_f = \frac{e^{0.0198}}{0.99} = 1.0303
$$

```python
R_f=np.exp(0.0198)/0.99
R_f
```

Define x as one plus dividend yield for market portfolio:

$$
x = \left(1 + \frac{P}{D}\right) \frac{D}{P} = 1 + \frac{D}{P}
$$

and define error term:

$$
e(x) = 0.99 b_0 \mathbb{E} [v(x \tilde{g})] + 0.99 x - 1
$$

where utility from recent financial gain or loss is given by:

$$
v(R) = R - 1.0303 \quad \text{for} \quad R \geq 1.0303
$$

$$
v(R) = 2 \left(R - 1.0303\right) \quad \text{for} \quad R < 1.0303
$$


```python
def v(R):
    return np.where(R >= R_f,
                    R-R_f,
                    2*(R-R_f)
                   )
def e_x(x,b0):
    R = x * g_tilde
    expected_v = np.mean(v(R))
    return 0.99 * b0 * expected_v + 0.99 * x - 1
```

```python
g_tilde.shape[0]
```

Solve for e(x) = 0 to find equilibrium value of x, using bisection search:

>1. Set x– = 1 and x+ = 1.1, and use simulated distribution of consumption growth to confirm that e(x–) < 0 and e(x+) > 0 ⇒ solution must lie between x– and x+

```python
def find_equilibrium_x(b0):
    # Initial bounds for bisection method
    x_lower, x_upper = 1, 1.1
    
    # Function to find root for e(x) = 0
    def target(x):
        return e_x(x, b0)
    
    # Use bisection method to find the root of e(x) = 0
    x_solution = optimize.bisect(target, x_lower, x_upper, xtol=1e-5) #error range
    return x_solution
```

>2. Set x0 = 0.5*(x– + x+) and use simulated distribution of consumption growth to calculate e(x0)

>3. If |e(x0)| < 10–5, then you have converged to solution
>
>4. Otherwise if e(x0) < 0, then solution lies between x0 and x+ ⇒ repeat from step 2 with x– = x0
>
>5. Otherwise if e(x0) > 0, then solution lies between x– and x0 ⇒ repeat from step 2 with x+ = x0

Repeat for b0 in range from 0 to 10, in increments of 0.1 (or less).

Calculate price-dividend ratio for market portfolio:

$$ \frac{P}{D}=\frac{1}{x-1}
$$

Plot price-dividend ratio (on vertical axis) vs b0.

Calculate expected market return:

$$ \mathbb{E}(\tilde{R_m})=\mathbb{E}(x \tilde{g})=xe^{0.0202}
$$

Plot equity premium (on vertical axis) vs b0.

```python
# Range of b0 values
b0_values = np.arange(0, 10.1, 0.1)
x_values = []
pd_ratios = []
expected_returns = []
equity_premiums = []

for b0 in b0_values:
    x = find_equilibrium_x(b0)
    x_values.append(x)
    
    # Price-Dividend Ratio
    pd_ratio = 1 / (x - 1)
    pd_ratios.append(pd_ratio)
    
    # Expected market return
    expected_return = np.mean(x * g_tilde)
    expected_returns.append(expected_return)
    
    # Equity premium
    equity_premium = expected_return - R_f
    equity_premiums.append(equity_premium)
    
results_df = pd.DataFrame({
    'b0': b0_values,
    'x': x_values,
    'Price-Dividend Ratio': pd_ratios,
    'Expected Market Return': expected_returns,
    'Equity Premium': equity_premiums
})

# 输出表格
results_df
```

```python
# Price-Dividend Ratio vs b0
plt.figure(figsize=(10, 6))
plt.plot(b0_values, pd_ratios, label='Price-Dividend Ratio')
plt.xlabel('b0')
plt.ylabel('Price-Dividend Ratio')
plt.title('Price-Dividend Ratio vs b0')
plt.legend()
plt.grid()
plt.show()

# Equity Premium  vs b0
plt.figure(figsize=(10, 6))
plt.plot(b0_values, equity_premiums, label='Equity Premium')
plt.xlabel('b0')
plt.ylabel('Equity Premium')
plt.title('Equity Premium vs b0')
plt.legend()
plt.grid()
plt.show()
```

Briefly describe (in words, without using mathematical equations or formulas) main characteristics of v(∙) as well as economic significance and implications of b0 and λ.

1. The function v(⋅) reflects the way investors feel about recent financial gains or losses. A key feature of this function is that it treats gains and losses very differently: when returns are high (above a certain threshold, like the risk-free rate), the utility increases steadily with returns. However, when returns drop below that threshold, the utility drops much more sharply. This is because losses are felt more intensely than equivalent gains—a behavior known as loss aversion. In simple terms, people feel the pain of losing money more strongly than the pleasure of making the same amount.

2. b0 determines extent to which utility from recent financial gain or loss contributes to investor's lifetime utility. A higher b0 makes investors more sensitive to financial ups and downs. So, an investor with a high b0 will react more strongly to changes in their portfolio, which could lead to behaviors like frequent trading or a tendency to react emotionally to market swings. In economic terms, b0 captures an investor’s emotional response to short-term performance.

3. λ is the coefficient of relative risk aversion for consumption shocks. A higher λ means that losses feel much more painful than equivalent gains feel pleasurable. This bias can push investors to choose safer assets or demand a higher premium for holding riskier ones, as they want compensation for the potential "pain" of losses. High values of λ help explain why investors might underinvest in stocks or why stocks carry an "equity premium"—a higher expected return to make up for the discomfort of risk.
