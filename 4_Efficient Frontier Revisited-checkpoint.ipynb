```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import statsmodels.api as sm
import warnings 
warnings.filterwarnings("ignore")
```

```python
df1=\
    pd.read_excel("Risk_Factors.xlsx")
df2=\
    pd.read_excel("Market_Portfolio.xlsx")
df3=\
    pd.read_excel("Industry_Portfolios.xlsx")
```

```python
Rm=df2["Market"]
```

>Estimate expected deviation from market return, for the ten industry portfolios:

$$R_i=E\left( \tilde{R}_i - \tilde{R}_m \right)$$

```python
for i in df3.columns[1:]:
    df1[f"Excess_{i}_market_return"]=df3[i]-Rm
print(df1.iloc[:,5:])
```

>Also estimate covariance matrix of return deviations, for the ten industry portfolios:

$$V_{ij}=Cov[(\tilde{R}_i - \tilde{R}_m ),(\tilde{R}_j-\tilde{R}_m)]
$$

```python
excess_returns=df1.iloc[:,5:]
cov_matrix=excess_returns.cov()
print(cov_matrix)
```

```python
mean_return=\
    excess_returns.mean(axis=0)
mean_return
```

```python
df_1=\
    df3.drop(columns="Date")
```

```python
R=mean_return.values
V=cov_matrix.values
e=np.ones(len(df_1.columns))
V_inv=np.linalg.inv(V)
```

$$ \alpha=R'V^{-1} e $$

```python
alpha=np.dot(np.dot(R.T,V_inv),e)
```

```python
alpha
```

$$ \zeta=R' V^{-1} R $$

```python
zeta=np.dot(np.dot(R.T,V_inv),R)
```

```python
zeta
```

$$ \delta=e' V^{-1} e $$

```python
delta=np.dot(np.dot(e.T,V_inv),e)
```

```python
delta
```

$$ Ïƒ_p^2=\frac{1}{\delta} +\frac{\delta}{\zeta \delta - \alpha^2}(R_p - R_{mv})^2 $$

```python
R_p=np.arange(0,0.105,0.005)
R_mv=alpha/delta
R_mv
```

```python
R_p
```

```python
sigma_p=\
    (1/delta)+(delta/(zeta*delta-alpha**2))*(R_p-R_mv)**2
sigma_p_squared=\
    np.sqrt(sigma_p)
```

```python
sigma_p_squared
```

>Plot the minimum-tracking-error frontier generated by the ten industry portfolios

```python
plt.plot(sigma_p_squared,R_p,label="minimum-tracking-error frontier")
plt.xlabel(r"(monthly) tracking error") 
plt.ylabel("expected (monthly) return deviation")
plt.title(r"minimum-tracking-error frontier")
plt.legend
plt.grid(True)

plt.show()
```

>Also plot the line starting from the origin that is tangent to the upper half of the minimum-tracking-error frontier.

$$ \sigma_p^2=\frac{(R_p-R_f)^2}{\zeta-2\alpha R_f+ \delta R_f^2} $$

```python
R_f=0
```

```python
sigma_p_sqr=(R_p-R_f)**2/(zeta-2*alpha*R_f+ delta*(R_f)**2)
# sigma_p_sqr
```

```python
sigma_p1=sigma_p_sqr**0.5
# sigma_p1
```

$$R_{p} = R_f + \sqrt{\left( \zeta - 2 \alpha R_f + \delta R_f^2 \right)}\sigma_{p}
$$

```python
R_p1=R_f+(zeta-2*alpha*R_f+delta*(R_f)**2)**0.5*sigma_p1
```

```python
plt.plot(sigma_p_squared,R_p,label="Minimum-tracking-error frontier")
plt.plot(sigma_p1, R_p1,label="Tangency")
plt.title("Efficient Minimum-Tracking-Error Frontier")
plt.xlabel(r"(monthly) tracking error") 
plt.ylabel("expected (monthly) return deviation")
plt.legend()
plt.grid(True)
plt.show()
```

>Calculate information ratio and portfolio weights for the "tangency" portfolio.

$$ 
\text{Information Ratio} = \left( \zeta - 2 \alpha R_f + \delta (R_f)^2 \right)^{\frac{1}{2}}
$$

```python
SP_ratio= (zeta-2*alpha*R_f+delta*(R_f)**2)**0.5
# SP_ratio
```

$$ \alpha=\frac{\zeta V^{-1}e-\alpha V^{-1}R}{\zeta \delta-\alpha^2} $$

```python
a=[zeta*np.dot(V_inv,e)-alpha*np.dot(V_inv,R)]/(zeta*delta-alpha**2)
# a
```

$$ b=\frac{\delta V^{-1}R-\alpha V^{-1}e}{\zeta\delta-\alpha^{2}} $$

```python
b=[delta*np.dot(V_inv,R)-alpha*np.dot(V_inv,e)]/(zeta*delta-alpha**2)
# b
```

$$ R_{tg}=\frac{\alpha R_f-\zeta}{\delta R_f-\alpha} $$

```python
R_tg=(alpha*R_f-zeta)/(delta*R_f-alpha)
# R_tg
```

$$ w^{*}=a + bR_p $$
$$ (R_p=R_{tg}) $$

```python
w_s=a+np.dot(b,R_tg)
# w_s
```

```python
pd.DataFrame(w_s.T,df_1.columns,columns=["Weights"])
```

Use Monte Carlo method to simulate the minimum-variance frontier without short sales, generated by the ten industry portfolios. Portfolio weights will be limited to the range [0, 1].

Randomly draw each element of w, the 10 Ã— 1 vector of portfolio weights, from the (standard) uniform distribution in the range [0, 1]. Divide w by the sum of portfolio weights, to ensure that the portfolio weights sum to one. This normalised w represents portfolio weights for one simulated portfolio, without short sales.

Use the normalised w along with the vector of mean returns and the covariance matrix of returns (for the ten industry portfolios) to calculate the mean return and standard deviation of return for the simulated portfolio. Repeat this process until you have (at least) 10^5 data points.

>Plot the data points with mean return on the vertical axis vs standard deviation of return on the horizontal axis.

$$R_p=w^T \cdot R$$
$${\sigma_p}^2=w^T \cdot \Sigma \cdot w$$

```python
mean_return1=\
    df_1.mean(axis=0)
cov_matrix1=\
    df_1.cov()
```

```python
target_data_points=1e5
portfolio_means=[]
portfolio_stds=[]
counts=0
while counts<target_data_points:
    w=np.random.uniform(0,1,len(mean_return1))
    w/=np.sum(w) #sum to one
    
    portfolio_return=np.dot(w,mean_return1)
    portfolio_variance=np.dot(np.dot(w.T,cov_matrix1),w)
    portfolio_std=np.sqrt(portfolio_variance)
    
    portfolio_means.append(portfolio_return)
    portfolio_stds.append(portfolio_std)
    
    counts += 1
    # if counts % 1 == 0:
    #     print(f"Generated {counts} portfolios")
        
plt.scatter(portfolio_stds,
            portfolio_means,
            marker="o",
            label="Simulated Portfolio")

plt.title('Monte Carlo Simulation of Minimum-Variance Frontier without Short Sales')
plt.xlabel('Portfolio Standard Deviation (Risk)')
plt.ylabel('Portfolio Mean Return')
plt.grid(True)
plt.legend()

plt.show()
```

Repeat this entire process by simulating 1/w using the standard uniform distribution â‡’ take the reciprocal of the random draw from the standard uniform distribution as the portfolio weight.

>Plot the new data points (on a separate graph) with mean return on the vertical axis vs standard deviation of return on the horizontal axis.

```python
target_data_points=1e5
portfolio_means=[]
portfolio_stds=[]
counts=0
while counts<target_data_points:
    w=np.random.uniform(0,1,len(mean_return1))
    w=1/w
    w/=np.sum(w) #sum to one
    
    portfolio_return=np.dot(w,mean_return1)
    portfolio_variance=np.dot(np.dot(w.T,cov_matrix1),w)
    portfolio_std=np.sqrt(portfolio_variance)
    
    portfolio_means.append(portfolio_return)
    portfolio_stds.append(portfolio_std)
    
    counts += 1
    # if counts % 1 == 0:
    #     print(f"Generated {counts} portfolios")
        
plt.scatter(portfolio_stds,
            portfolio_means,
            marker="o",
            label="Simulated Portfolio")

plt.title('Monte Carlo Simulation of Minimum-Variance Frontier with 1/w Weights')
plt.xlabel('Portfolio Standard Deviation (Risk)')
plt.ylabel('Portfolio Mean Return')
plt.grid(True)
plt.legend()

plt.show()
```
